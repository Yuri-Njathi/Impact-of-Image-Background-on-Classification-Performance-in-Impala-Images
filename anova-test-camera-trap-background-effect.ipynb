```python
import numpy as np
import scipy.stats as stats
import statsmodels.api as sm
from statsmodels.formula.api import ols
import pandas as pd

# Example data
data = {
    'background': ['A', 'A', 'A', 'A', 'A', 'B', 'B', 'B', 'B', 'B', 'C', 'C', 'C', 'C', 'C'],
    'accuracy': [0.85, 0.80, 0.82, 0.84, 0.83, 0.75, 0.78, 0.74, 0.77, 0.76, 0.90, 0.88, 0.91, 0.89, 0.92]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Fit the model
model = ols('accuracy ~ C(background)', data=df).fit()

# Perform ANOVA
anova_table = sm.stats.anova_lm(model, typ=2)

print(anova_table)
```

```python
import scipy.stats as stats

# Example data
background_A = [0.85, 0.80, 0.82, 0.84, 0.83]
background_B = [0.75, 0.78, 0.74, 0.77, 0.76]
background_C = [0.90, 0.88, 0.91, 0.89, 0.92]

# Perform one-way ANOVA
f_statistic, p_value = stats.f_oneway(background_A, background_B, background_C)

print(f"F-statistic: {f_statistic}, P-value: {p_value}")
```
